What can you observe on `df_airbnb_part2`. Look at the dataset. It is not too large !

Your answer: ....

Write a code to clean / delete the messy data. The new cleaned dataframe will be called `df_airbnb_part2_clean`. Tips: think about using indexing/slicing ;)

```python
# Start writing code here...

from sklearn import preprocessing

x = df_airbnb_part2["Price"].values.reshape(-1, 1)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df_airbnb_part2['Price'] = pd.DataFrame(x_scaled)

df_airbnb_part2
```

## 4. Combine the dataset

The file, being in 2 parts. We would like to combine them into 1 file. Is this a `concat` ou `merge` operation ? Do the right operation looking at the documentation :
- `concat` : https://pandas.pydata.org/docs/reference/api/pandas.concat.html
- `merge` : https://pandas.pydata.org/docs/reference/api/pandas.merge.html

We will call the obtained table: `df_airbnb`

```python
# Start writing code here...

data = [df_airbnb_part1, df_airbnb_part2.iloc[:7]]
df_airbnb = pd.concat(data)
df_airbnb.reset_index(drop=True, inplace=True)
total_calculation = pd.concat([df_airbnb_part1, df_airbnb_part2]).sum()
total_calculation_df = pd.DataFrame(total_calculation).T
df_airbnb = pd.concat([df_airbnb, total_calculation_df])
df_airbnb.reset_index(drop=True, inplace=True)
df_airbnb.tail(10)
```

## 8. Get Insights

What is the average price in NYC ? Store this price in a variable called `mean_price` and print it into the console.

```python
# Start writing code here...
avg_price = df_airbnb['Price'].mean()
print("Average price in NYC: ", avg_price,"$")
```

What is the total number of reviews in NYC ? Store this total reviews in a variable called `total_reviews` and print it into the console.

Tips: Give a nice print ;)

```python
# Start writing code here...

valid_reviews = df_airbnb['Review Scores Rating'].notna()
total_reviews = valid_reviews.sum()
print("Total number of reviews in NYC:", total_reviews)
```

What are the Top 3 most representated Property Type of the hosts ? Store it in a variable called `top3_property_type` and display only the Top 3

```python
# Start writing code here...

top3_property_type = df_airbnb['Property Type'].value_counts().head(3)
print("Top 3 most represented property types:")
print(top3_property_type)
```

What is the less representated Room Type among the hosts ? Store it in a variable called `bottom_room_type` and display only this room type

```python
# Start writing code here...

bottom_room_type = df_airbnb['Property Type'].value_counts().tail(3)
print("Top 3 most represented property types:")
print(bottom_room_type)
```
